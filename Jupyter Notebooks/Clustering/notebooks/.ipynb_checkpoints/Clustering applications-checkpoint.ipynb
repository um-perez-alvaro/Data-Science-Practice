# Clustering Applications

## Finding patterns in data

Data from [here](https://archive.ics.uci.edu/ml/datasets/Metro+Interstate+Traffic+Volume)

```python
data = pd.read_csv('metro.csv', index_col='date_time', parse_dates=True)
data.head()
```

```python
data = data[data.index.year>=2016]
```

```python
data.traffic_volume.plot(figsize=(20,5))
```

```python
table = pd.pivot_table(data=data,values='traffic_volume', index=data.index.time, columns = data.index.date)
```

```python
table.plot(figsize=(10,10), legend=False, alpha = 0.2)
```

```python
table
```

```python
(
    table
    .isnull()
    .sum(axis=0)
    .sort_values(ascending=False)
    .head(10)
)
```

```python
table.interpolate().head()#.fillna(method = 'bfill')
```

```python
table = table.interpolate().fillna(method = 'bfill')
```

```python
from sklearn.cluster import KMeans
k=2
kmeans = KMeans(n_clusters=k)
kmeans.fit(table.T)
```

```python
# cluster labels and centers
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# put cluster centers into a dataframe
cluster_centers = pd.DataFrame(centers.T, index=table.index)
```

```python
# plot cluster centers + clusters
plt.figure(figsize=(20,20))
for i in range(k):
    plt.subplot(k,1,i+1)
    ax = plt.gca()
    table.iloc[:,labels==i].plot(ax=ax, legend=False,alpha=0.2)
    cluster_centers[i].plot(ax = ax, color='red', linewidth=7)    
```
