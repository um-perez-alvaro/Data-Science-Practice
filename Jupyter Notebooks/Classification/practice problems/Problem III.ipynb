# The MNIST dataset: Classification of handwritten digits

```python
# load the MNIST data
path = 'https://raw.githubusercontent.com/um-perez-alvaro/Data-Science-Practice/master/Data/digits.csv'
df = pd.read_csv(path)
df.head()
```

```python
# feature matrix and target vector
X = df.iloc[:,0:784]
y = df['class']
```

There are 5,000 images in the dataset, each represented by 784 features. This is because each image is 28×28 pixels, and each feature corresponds to the intensity of a pixel, ranging from 0 (white) to 255 (black).

Let’s take a look at one digit from the dataset.
We’ll grab a feature vector (a row from X), reshape it into a 28x28 array, and display it using `plt.imshow()`.

```python
some_digit = X.loc[0,:].values
some_digit_image = some_digit.reshape(28,28)
plt.imshow(some_digit_image,cmap = 'binary')
plt.axis('off')
plt.show()
```

It looks like a 5, and that’s exactly what the label confirms.

```python
plt.figure(figsize=(10,10))
for i in range(100):
    digit = X.loc[i,:].to_numpy()
    digit_image = digit.reshape(28,28)
    plt.subplot(10,10,i+1)
    plt.imshow(digit_image,cmap = 'binary')
    plt.axis('off')
```
